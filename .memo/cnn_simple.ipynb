## フェーズ0: 超シンプルCNNでの動作確認 (`feature/simple_cnn_loop`)

構築したデータパイプラインと、新しく作成した学習モデル・学習ロジックを接続し、データ入力からモデルの重み更新まで、一連の学習フローがエラーなく実行できることを確認しました。

### やったこと

- **単純なCNNモデルの実装 (`src/model.py`)**
  - 学習ループ全体の動作確認を目的として、2つの畳み込み層と2つの全結合層からなる非常にシンプルなCNNモデル (`SimpleCNN`) を作成しました。
  - これにより、複雑なモデルを導入する前に、基本的なアーキテクチャでパイプラインが機能することを検証できます。

- **基本的な学習スクリプトの作成 (`src/train.py`)**
  - データローダーからバッチ単位で画像とラベルを取得します。
  - モデルの出力と正解ラベルから損失（交差エントロピー誤差）を計算します。
  - 計算した損失に基づいて勾配を計算し、オプティマイザ（Adam）を使ってモデルのパラメータを更新します。
  - GPUが利用可能な場合は自動で `cuda` を使用し、高速な学習を行えるようにしました。

- **学習フローの全体検証**
  - `dataset.py`、`model.py`、`train.py`を連携させ、学習が正常に開始し、エポックごとに損失が表示されることを確認しました。

### 動作確認

構築したモデルと学習スクリプトは、`src/train.py` を直接実行することで動作確認が可能です。

```bash
# srcディレクトリ内から実行
python3 train.py
```
実行すると、エポックごとの学習損失が出力され、学習ループが正常に機能していることを確認できます。
