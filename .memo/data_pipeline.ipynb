
## フェーズ0: データパイプラインの構築 (`feature/data_pipeline`)

多様な形式の画像データをPyTorchが学習可能な、均質で正規化されたテンソル形式に変換するための基盤を構築しました。

### やったこと

-   **カスタムデータセットクラスの実装 (`src/dataset.py`)**
    -   `torch.utils.data.Dataset` を継承した `HumanCharacterDataset` クラスを作成しました。
    -   これにより、プロジェクト固有のデータ読み込みロジックをカプセル化しました。

-   **訓練・テストデータの自動分割**
    -   「ファイル名の末尾3桁の連番が4の倍数のものをテストデータとする」というルールに基づき、データセットを訓練用とテスト用に自動で振り分ける機能を実装しました。
    -   `01-001.jpg` のようなハイフンを含むファイル名にも対応する、頑健なファイル番号の解析処理を実装しました。

-   **画像前処理パイプラインの構築**
    -   `torchvision.transforms` を利用し、以下の処理をパイプライン化しました。
        1.  **リサイズ**: 全ての画像のサイズを `224x224` に統一し、モデルへの入力サイズを揃えました。
        2.  **テンソル変換**: PIL形式の画像をPyTorchのテンソルに変換しました。
        3.  **正規化**: ImageNetの平均と標準偏差を用いてピクセル値を正規化し、転移学習の準備を整えました。
